```python
#installing pandas and cv lib
#image is an 2-D dimensional array consisting values in pixels.
# Each color in computer is stored in RGB form while in openCV, color is stored in BGR format.

import cv2
import pandas as pd

imag_path = 'pic1.jpg'
csv_path = 'colors.csv'

# reading csv files.
# color,color_name,hex_value,R,G,B.

index = ['color', 'color_name', 'hex', 'R', 'G', 'B']
df = pd.read_csv(csv_path, names=index, header=None)

# reading image
#it is displayed using window.
#resizing the image on the window.
imag = cv2.imread(imag_path)
imag = cv2.resize(imag, (800,600))

#declaring global variables
clicked = False # which means it is not clicked anywhere now.
r = g = b = x_val = y_val = 0

#function to calculate minimum distance from all colors and get the most matching color
# finding by absolute differences of each value with the value stored in csv file.
#so,iterating through csv and where we get the minimum abs. diff (d) that will be the output.
def get_color_name(R,G,B):
    minm = 1000
    for i in range(len(df)):
        d = abs(R - int(df.loc[i,'R'])) + abs(G - int(df.loc[i,'G'])) + abs(B - int(df.loc[i,'B']))
        if d <= minm:
            minm = d
            col_name = df.loc[i, 'color_name']

    return col_name

# finding the x and y coordinates of the mouse pointer so that we get to know the exact location of the click on image.
# event is double-click,left-click,right-click and so on.
#binding this function with mouse pointer.
def draw_function(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        global b, g, r, x_val, y_val, clicked # pass by refrence not possible in python.so,global
        clicked = True
        x_val = x
        y_val = y
        b,g,r = imag[y,x] # RGB in cv is the form of b,g,r.
        b = int(b) # b,g,r in numpy value.so,coverting it.
        g = int(g)
        r = int(r)

# creating window and passing names in ()
cv2.namedWindow('image')
#calling method to set above functin with mouse click.
cv2.setMouseCallback('image', draw_function)

while True: # to keep programming running until escape is clicked.
    cv2.imshow('image', imag)
    if clicked:
        #cv2.rectangle(image, startpoint, endpoint, color, thickness)-1 fills entire rectangle 
        cv2.rectangle(imag, (20,20), (600,60), (b,g,r), -1)
        
        #Creating text string to display( Color name and RGB values )
        text = get_color_name(r,g,b) + ' R=' + str(r) + ' G=' + str(g) + ' B=' + str(b)
        #cv2.putText(img,text,start,font(0-7),fontScale,color,thickness,lineType )
        cv2.putText(imag, text, (50,50), 2,0.8, (255,255,255),2,cv2.LINE_AA)

        #For very light colours we will display text in black colour
        if r+g+b >=600:
            cv2.putText(imag, text, (50,50), 2,0.8, (0,0,0),2,cv2.LINE_AA)
    
    # means until and unless we do not click the cross button of window,the window will not close.
    if cv2.waitKey(20) & 0xFF == 27: #escape key
        break

cv2.destroyAllWindows() # when cross button is clicked then the windows is destroyed.
```
